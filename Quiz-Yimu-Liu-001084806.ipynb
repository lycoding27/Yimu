```python
for n in range(1,92):
    
    data = pd.read_csv('book1.csv')
    text = data.values[n-1][0]
    TINT_COLOR = (0, 0, 0)  # Black
    TRANSPARENCY = .28  # Degree of transparency, 0-100%
    OPACITY = int(300 * TRANSPARENCY)
    
    from PIL import Image, ImageFont, ImageDraw
    
    sa='MonkeyKing/A%d.png' %n
    sb='MonkeyKing/B%d.png' %n
    
    img = pili.open(sa).convert('RGBA')
    #img = pili.alpha_composite(img, overlay)
    img = img.convert("RGB") # Remove alpha for saving in jpg format.
    img.save(sb)
    
    piliops.expand(pili.open(sb), border=(5,5),fill='white').save(sb)
    #img = pili.open('images/A%r.jpg' %(n)).convert('RGBA')
    img = pili.open(sb).convert('RGBA')
    overlay = pili.new('RGBA', img.size, TINT_COLOR+(0,))
    draw = pild.Draw(overlay)
    font = pilf.truetype("方正兰亭黑简.TTF", 28)
    
    w, h = font.getsize(text)
    num_lines = len(text.split('\n'))
    print(num_lines, w, h)
    x, y = 10, img.height - (num_lines-0.05*num_lines)*h
    draw.rectangle((x, y, x + img.width - 200, y + (num_lines-0.1*num_lines)*h), fill=TINT_COLOR+(OPACITY,))
    draw.text((x, y), text, fill=(209, 239, 8), font=font)
    
    img = pili.alpha_composite(img, overlay)
    img = img.convert("RGB") # Remove alpha for saving in jpg format.
    sc='MonkeyKing/C%d.png' %n
    img.save(sc) 
    
```

```python
import numpy as np
import PIL
list_im = ['MonkeyKing/C1.png',
            'MonkeyKing/C2.png',
            'MonkeyKing/C3.png',
            'MonkeyKing/C4.png',
            'MonkeyKing/C5.png',
          'MonkeyKing/C6.png',
          'MonkeyKing/C7.png',
          'MonkeyKing/C8.png',
          'MonkeyKing/C9.png',
          'MonkeyKing/C10.png']
imgs    = [ PIL.Image.open(i) for i in list_im ]

# pick the image which is the smallest, and resize the others to match it (can be arbitrary image shape here)
min_shape = sorted( [(np.sum(i.size), i.size ) for i in imgs])[0][1]
imgs_comb = np.vstack( (np.asarray( i.resize(min_shape) ) for i in imgs ) )

# save that beautiful picture
imgs_comb = PIL.Image.fromarray( imgs_comb)
imgs_comb.save( 'MonkeyKing/P10.png' ) 
```

# From images to pdfs


```python
from PIL import Image
from fpdf import FPDF

#pdf = FPDF()
cover = Image.open('MonkeyKing/test.png')
width, height = cover.size
pdf = FPDF(unit = "pt", format = [width, height])
    
# imagelist is the list with all image filenames
imagelist = ['MonkeyKing/P100.png']

for image in imagelist:
    pdf.add_page()
    pdf.image(image, 0, 0, width, height)
pdf.output("MonkeyKing/Finalll.pdf", "F")
print("done!")
```
